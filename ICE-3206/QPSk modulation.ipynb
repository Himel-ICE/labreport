```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(0, 2, 10)
x
```

```python
p = np.where(x == 0, -1, 1)
p
```

```python
even_seq = p[::2]
odd_seq = p[1::2]
print(even_seq)
print(odd_seq)
```

```python
t = np.arange(0, len(x), 0.01)
len(t)
```

```python
even_ps = np.repeat(even_seq, int(len(t) / len(even_seq)))
odd_ps = np.repeat(odd_seq, int(len(t) / len(odd_seq)))
```

```python
c1 = np.cos(2 * np.pi * 1 * t)
c2 = np.sin(2 * np.pi * 1 * t)
```

```python
r1 = even_ps * c1
r2 = odd_ps * c2
qpsk_sig = r1 - r2
```

```python
plt.figure(figsize=(10, 3))
plt.plot(t, even_ps)
plt.grid(True)
plt.title("Even sequence")
```

```python
plt.figure(figsize=(10, 3))
plt.plot(t, odd_ps)
plt.grid(True)
plt.title("ODD sequence")
```

```python
plt.figure(figsize=(10, 3))
plt.plot(t, c1)
plt.grid(True)
```

```python
plt.figure(figsize=(10, 3))
plt.plot(t, c2)
plt.grid(True)
```

```python
plt.figure(figsize=(10,3))
plt.plot(t, r1)
plt.grid(True)
```

```python
plt.figure(figsize=(10, 3))
plt.plot(t, r2)
plt.grid(True)
```

```python
plt.figure(figsize=(10, 3))
plt.plot(t, qpsk_sig)
plt.grid(True)
```
