```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10
n = np.random.randint(0, 2, N)
n
```

```python
nnn = []
for m in range(N):
    if n[m] == 1:
        nn = [1, -1]
    else:
        nn = [-1, 1]
    nnn.extend(nn)
nnn
```

```python
i = 0
l = 0.5
t = np.arange(0, N, 0.01)
t[: 200]
```

```python
y = []
for j in range(len(t)):
    if t[j] <= l:
        y.append(nnn[i])
    else:
        y.append(nnn[i])
        i = i + 1
        l = l + 0.5
#y[800:1000]
```

```python
plt.figure(figsize=(10, 3))
plt.plot(t, y)
plt.grid(True)
```
