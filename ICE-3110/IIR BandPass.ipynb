# IIR Bandpass

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
```

```python
fs = 8000
[n, w] = sig.buttord([1000 / 4000, 2500 / 4000], [400 / 4000, 3200 / 4000], 1, 50)
# "n" is filter order and "w" is critical frequency
[b, a] = sig.butter(n, w, btype = 'bandpass')
w, h = sig.freqz(b, a, 512, fs = fs)
z, p, k = sig.tf2zpk(b, a)
```

```python
plt.figure(figsize = (9, 4))
plt.plot(w, np.abs(h))
plt.grid(True)
```

```python
plt.figure(figsize = (9, 4))
plt.plot(w, np.unwrap(np.angle(h)))
plt.grid(True)
```

```python
plt.figure(figsize = (10, 6))
plt.scatter(np.real(z), np.imag(z), marker='o', edgecolors='black')
plt.scatter(np.real(p), np.imag(p), marker='x', color='b')
plt.grid(True)
```
