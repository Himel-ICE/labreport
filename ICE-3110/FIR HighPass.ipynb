# FIR Highpass filter

```python
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt
```

```python
fs = 4000 
N = 100
fc = 1200
b = sig.firwin(N + 1, fc, fs = fs, window = 'hamming', pass_zero = 'highpass')
w, h_freq = sig.freqz(b, fs = fs)
z, p, k = sig.tf2zpk(b, 1)
```

```python
plt.figure(figsize = (8, 4))
plt.title('Impulse response of FIR filter')
plt.plot(np.arange(len(b)), b)
plt.xlabel('sample')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (10, 2))
plt.plot(w, np.abs(h_freq))
plt.grid(True)
plt.xlabel('frequency(Hz)')
plt.ylabel('Magnitute')
```

```python
plt.figure(figsize = (10, 3))
plt.plot(w, np.unwrap(np.angle(h_freq)))
plt.xlabel('frequency(Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
```

```python
plt.figure(figsize = (9, 4))
plt.scatter(np.real(z), np.imag(z), marker = 'o', edgecolors = 'green')
plt.scatter(np.real(p), np.imag(p), marker = 'x', edgecolors = 'black')
plt.xlabel('real')
plt.ylabel('imaginary')
plt.grid(True)
plt.show()
```
