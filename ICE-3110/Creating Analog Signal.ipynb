# FFT

```python
import numpy as np
import matplotlib.pyplot as plt
```

#to perform DFT

```python
n = np.arange(-1, 4)
x = np.arange(1, 6)
N = len(n)
k = np.arange(N)
```

```python
X = np.sum(x * np.exp(-2j * np.pi * np.outer(n, k) / N), axis = 1)
```

```python
plt.figure(figsize = (4, 4))
plt.plot(k, np.abs(X))
plt.grid(True)
plt.xlabel('frequency in pi units')
plt.title('Magnitude part')
```

```python
plt.figure(figsize = (4, 4))
plt.plot(k, np.angle(X))
plt.grid(True)
plt.xlabel('frequency in pi units')
plt.title('Angle part')
```

```python
plt.figure(figsize = (4, 4))
plt.plot(k, np.real(X))
plt.grid(True)
plt.xlabel('frequency in pi units')
plt.title('Real part')
```

```python
plt.figure(figsize = (4, 4))
plt.plot(k, np.imag(X))
plt.grid(True)
plt.xlabel('frequency in pi units')
plt.title('Imaginary part')
```

```python
fs = 128
N = 256
T = 1 / fs
k = np.arange(N)
time = k * T
f =.25 + 2 * np.sin(2 * np.pi * 5 * k * T) + 1 * np.sin(2 * np.pi * 12.5 * k * T) + 1.5 * np.sin(
    2 * np.pi * 20 * k * T) + 0.5 * np.sin(2 * np.pi * 35 * k * T)
```

```python
plt.figure(figsize = (12, 3))
plt.plot(time, f)
plt.grid(True)
```

```python
F = np.fft.fft(f)
magF = np.abs(np.hstack((F[0] / N, F[1 : N // 2] / (N / 2))))
hertz = k[0 : N // 2] * (1 / (N * T))
```

```python
plt.figure(figsize = (10, 5))
plt.stem(hertz, magF)
plt.title('Frequency Components')
```
