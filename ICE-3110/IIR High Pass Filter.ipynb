# IIR Low Pass

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
```

```python
fs = 8000
[n, w] = sig.buttord(1200 / 4000, 2000 / 4000, 1, 50)
[b, a] = sig.butter(n, w, btype = 'highpass')
w, h = sig.freqz(b, a, 512, fs = fs)
z, p, k = sig.tf2zpk(b, a)
```

```python
plt.figure(figsize   = (10, 4))
plt.plot(w, np.abs(h))
plt.grid(True)
```

```python
plt.figure(figsize   = (10, 4))
plt.plot(w, np.unwrap(np.angle(h)))
plt.grid(True)
```

```python
plt.figure(figsize = (10, 6))
plt.scatter(np.real(z), np.imag(z), marker = 'o', edgecolors = 'black')
plt.scatter(np.real(p), np.imag(p), marker = 'x', edgecolors = 'red')
plt.grid(True)
```
