# Sampling

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = 5 # frequency
fs = 100 #sampling frequency
t = np.arange(0, 1, 1 / fs)
x = np.sin(2 * np.pi * f * t)
```

```python
plt.figure(figsize = (10, 4))
plt.plot(t, x)
plt.xlabel('time(s)')
plt.ylabel('Amplitude')
plt.grid(True)
```

```python
ts = 1 / fs
n = np.arange(0, 1, ts)
xn = np.sin(2 * np.pi * f * n)
```

```python
plt.figure(figsize = (10, 4))
plt.stem(n, xn)
plt.xlabel('time(s)')
plt.ylabel('Amplitude')
plt.grid(True)
```

```python
xr = np.zeros_like(t)
for i in range(len(n)):
    xr += xn[i] * np.sinc((t - i * ts) / ts)
```

```python
plt.figure(figsize = (10, 4))
plt.plot(t, xr)
plt.grid(True)
plt.xlabel('time(s)')
plt.ylabel('Amplitude')
```
