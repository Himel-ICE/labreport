# DFT and IDFT

```python
import numpy as np
import matplotlib.pyplot as plt
```

# DFT

```python
signal = [0, 1, 0, 1]
```

```python
dft_result = np.fft.fft(signal)
dft_result
```

```python
mag = np.abs(dft_result)
freq = np.fft.fftfreq(len(signal))
```

```python
plt.stem(freq, mag)
plt.grid(True)
plt.xlabel("Frequency(Hz)")
plt.ylabel("Magnitude")
plt.title("DFT")
```

# IDFT

```python
idft_result = np.fft.ifft(dft_result)
idft_result
```

```python
plt.stem(np.arange(len(idft_result)), np.real(idft_result))
plt.grid(True)
plt.xlabel("Points")
plt.ylabel("Amplitude")
plt.title("IDFT")
```
