# FIR MultiBandPass Filter

```python
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt
```

```python
fs = 8000
N = 100
fc = np.array([1200, 1800, 2600, 3200])
```

```python
b = sig.firwin(N + 1, fc, fs = fs, window = 'hamming', pass_zero = 'bandpass')
```

```python
plt.figure(figsize = (8, 4))
plt.stem(np.arange(len(b)), b)
plt.grid(True)
plt.xlabel('sample')
plt.ylabel('Amplitude')
```

```python
w, h_freq = sig.freqz(b, fs = fs)
```

```python
plt.figure(figsize = (9, 3))
plt.plot(w, np.abs(h_freq))
plt.grid(True)
plt.xlabel('frequency(Hz)')
plt.ylabel('Amplitude')
```

```python
plt.figure(figsize = (9, 3))
plt.plot(w, np.unwrap(np.angle(h_freq)))
plt.grid(True)
plt.xlabel('frequency(Hz)')
plt.ylabel('Amplitude')
```

```python
z, p, k = sig.tf2zpk(b, 1)
```

```python
plt.figure(figsize = (9, 4))
plt.scatter(np.real(z), np.imag(z), marker = 'o', edgecolors = 'black')
plt.grid(True)
plt.xlabel('Real')
plt.ylabel('Imaginary')
```
