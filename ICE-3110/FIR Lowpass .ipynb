# FIR lowpass filter

```python
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt
```

```python
fs = 8000 #sampling range
N = 100 #order of filter
fc = 500 # cutoff frequency
b = sig.firwin(N + 1, fc, fs = fs, window = 'hamming', pass_zero = 'lowpass') # coefficient of the filter
w, h_freq =  sig.freqz(b, fs)
z, p, k = sig.tf2zpk(b, 1)
```

```python
plt.figure(figsize = (8, 4))
plt.title('Impulse response of FIR filter')
plt.stem(np.arange(len(b)), b)
plt.xlabel('sample')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (8, 2))
plt.plot(w, np.abs(h_freq)) #magnitude
plt.grid(True)
plt.xlabel('frequency(kHz)')
plt.ylabel('Magnitude')
```

```python
plt.figure(figsize = (8, 2))
plt.plot(w, np.unwrap(np.angle(h_freq))) # phase
plt.grid(True)
plt.xlabel('frequency(kHz)')
plt.ylabel('Magnitude')
```

```python
plt.figure(figsize = (9, 4))
plt.scatter(np.real(z), np.imag(z), marker = 'o', edgecolors = 'b')
plt.scatter(np.real(p), np.imag(p), marker = 'x', color = 'b')
plt.xlabel('real')
plt.ylabel('imaginary')
plt.grid(True)
plt.show()
```
